```python
# Import dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month

y_axis= [10.02, 23.34, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
    #plt.plot - creates figure w/ the current axes
    
plt.plot(x_axis,y_axis)
```

```python
# create a plot w/ ax.plot()

#create the figure
fig, ax = plt.subplots()


# define the axes
ax.plot(x_axis, y_axis)
```

```python
# Breaking down the code

    #1 plt.subplots() - returns a tuple that contains a figure & axes obj
    #2 fig, ax = plt.subplots() - unpacks the tuple into the variables fig and ax.
    #3 The first variable, fig, references the figure.
    #4 The second variable, ax, references the axes of the figure and allows us to annotate the figure.
    #5 The ax.plot(x_axis, y_axis) plots the data from the x and y axes.
            # can also be written
                # Create the plot with ax.plt()
                # fig = plt.figure()
                # ax = fig.add_subplot()
```

```python
# Using plt.show() - looks for all active fig objects, opens window and displays them
    
    # can only be used once/session, may need to restart kernel


plt.plot(x_axis, y_axis)

plt.show()
```

```python
# Create the plot & add labels for the legend

plt.plot(x_axis, y_axis, label= 'Boston')

# Create labels for the x and y axes 

plt.xlabel("Date")

plt.ylabel("Fare($)")

# set the y limit btwn 0 and 45

plt.ylim(0, 45)

# create title

plt.title("PyBer Fare by Month")

# add legend

plt.legend()
           
```

```python
# Pie Charts using matplotlib

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09] 

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09] 


colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots()
ax.pie(y_axis, 
       explode= explode_values, 
       labels= x_axis, 
       colors=colors, 
       autopct='%.1f%%', 
       shadow=True, 
       startangle=90, 
       radius=2,
       counterclock=False)
plt.show()
```
